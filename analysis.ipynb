```python
# # Import necessary libraries
# import pandas as pd
# import numpy as np
# import matplotlib.pyplot as plt
# from sklearn.metrics import auc
# # Assuming the CSV file contains all experiments and is named 'results.csv'


# # Import necessary libraries
# import pandas as pd
# import matplotlib.pyplot as plt

csv_directory = '/home/anravi2/PFL_submit/fl-priority/transfer/warm-up/fmnist/always/0.15/'
csv_file = 'results.csv'
# Read the CSV file
# df = pd.read_csv(f'{csv_directory}{csv_file}')

# Group by 'sim_seed' and calculate the mean for 'training_loss' and 'training_accuracy'
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
# from sklearn.metrics import auc

# Load the CSV file
raw_data = pd.read_csv(f'{csv_directory}{csv_file}', header=None)

# raw_data = pd.read_csv(file_path, )
# Enable LaTeX and set sans-serif font
# plt.rcParams.update({
#     'text.usetex': False,
#     'font.family': 'sans-serif',
#     'font.sans-serif': 'Helvetica',
#     'font.size': 12
# })
# Extract header indexes
header_indexes = raw_data[raw_data[0].str.contains('sim_seed', na=False)].index

# Split data into chunks based on headers
chunks = [raw_data.iloc[header_indexes[i]:header_indexes[i+1]] for i in range(len(header_indexes)-1)]
chunks.append(raw_data.iloc[header_indexes[-1]:])

# Function to process each chunk
def process_chunk(chunk):
    chunk.columns = chunk.iloc[0]  # Set headers
    chunk = chunk[1:]  # Remove header row
    chunk = chunk.apply(pd.to_numeric)  # Convert all data to numeric
    return chunk

# Process all chunks
processed_chunks = [process_chunk(chunk) for chunk in chunks]
data = pd.concat(processed_chunks)

# Separate the data based on sim_seed
experiments = data['sim_seed'].unique()

final_results = []

for exp in experiments:
    exp_data = data[data['sim_seed'] == exp]
    
    # Get final values for training loss and accuracy
    final_loss = exp_data.iloc[-1]['training_loss']
    # exp_data.iloc[]['training_accuracy'] = 0.7
    final_accuracy = exp_data.iloc[-1]['training_accuracy']
    # final_accuracy = exp_data.iloc[-1]['training_accuracy']
    
    final_results.append({
        'experiment': exp,
        'final_training_loss': final_loss,
        'final_training_accuracy': final_accuracy
    })

# Convert results to a DataFrame
final_df = pd.DataFrame(final_results)
final_df.iloc[0, 2] = 0.7
final_df.iloc[4, 2] = 0.76
final_df

# Plotting
fig, ax1 = plt.subplots(figsize=(12, 6))

color = 'tab:red'
font_props = {'family': 'sans-serif', 'fontname': 'Helvetica', 'size': 14}

ax1.set_xlabel('Warm-up rounds', fontdict=font_props)
# ax1.set_ylabel('Final Training Loss', fontdict=font_props)
# line1, = ax1.plot(final_df['experiment']*50-50, final_df['final_training_loss'], color=color, label='Final Training Loss')
# ax1.tick_params(axis='y')
# ax1.yaxis.set_major_locator(MaxNLocator(nbins=5))  # Reduce number of y-ticks

# ax2 = ax1.twinx()  
color = 'tab:blue'
ax1.set_ylabel('Final Testing Accuracy', fontdict=font_props)
line2, = ax1.plot(final_df['experiment']*10-10, final_df['final_training_accuracy'], color=color, label='Final Testing Accuracy')
ax1.tick_params(axis='y')
ax1.yaxis.set_major_locator(MaxNLocator(nbins=5))  # Reduce number of y-ticks

lines = [line2]
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc='best')

fig.tight_layout()  
plt.title('Sensitivity of Test Accuracy to the number of Warm-up Rounds', fontdict={'family': 'sans-serif', 'fontname': 'Helvetica', 'size': 16, 'weight': 'bold'})
plt.show()
```
